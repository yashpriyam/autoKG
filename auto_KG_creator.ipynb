```python
from tkinter import *
import pyautogui, time
from splinter import Browser
from selenium.webdriver.common.keys import Keys
import xlrd
import keyboard


window=Tk()
window.configure(background='SeaGreen1')

def open_me():
    global b
    b=Browser("chrome")
    url = "http://52.77.223.93/user/login"
    b.visit(url)

    b.find_by_id('id_username').first.fill(e1_value.get())
    b.find_by_id('id_password').first.fill(e2_value.get())


    b.find_by_value('login').first.click()
    b.find_by_css('.home-nav').first.click()
    b.click_link_by_partial_href('/graphs/edit/new')

    time.sleep(5)

    loc = (e3_value.get()) 
    wb = xlrd.open_workbook(loc) 
    sheet = wb.sheet_by_index(0)
    l = []
    for i in range(sheet.nrows):
        d = {}
        d["Concept"] = sheet.cell_value(i, 5)
        d["Description"] = sheet.cell_value(i, 6)
        l.append(d)
    res = [lub['Concept'] for lub in l]
    mes = [dub['Description'] for dub in l]

   
    pyautogui.click(x=300, y=500, clicks=1, button='left')
    time.sleep(2)

    i = 0
    while i < len(res) and i < len(mes):    
        pyautogui.keyDown('shift')
        time.sleep(0.2)
        pyautogui.click(50, 500, clicks=1, button='left')
        time.sleep(0.2)
        pyautogui.keyUp('shift')
        time.sleep(0.2)
        pyautogui.press('backspace')
        time.sleep(0.2)
        pyautogui.press('enter')
        time.sleep(0.5)
        pyautogui.click(77, 473, clicks=1, button='left')
        time.sleep(0.5)
        pyautogui.click(x=350, y=200, clicks=1, button='left')
        time.sleep(0.2)
        pyautogui.typewrite(res[i])
        time.sleep(0.2)
        pyautogui.click(x=350, y=350, clicks=1, button='left')
        time.sleep(0.2)
        pyautogui.typewrite(mes[i])
        time.sleep(0.5)
        pyautogui.click(50, 500, clicks=2, button='left')
        time.sleep(2)
        b.reload()
        time.sleep(2)
        if keyboard.is_pressed('q'):
            break
        else:
            i += 1

    link=(b.url)
    t1.delete("1.0", END)
    t1.insert(END,link)
    
t1=Text(window,height=1,width=42)
t1.grid(row=10,column=1,padx=5,pady=5)
t2=Label(window,text="Link to your metacademy graph")
t2.grid(row=9,column=1,padx=5,pady=5)

b1=Button(window,text="Start",command=open_me,background='IndianRed1',foreground='black',activebackground='green2')
b1.grid(row=11,column=1,padx=10,pady=10)

e1_value=StringVar()
e1=Entry(window,width=20,textvariable=e1_value,highlightcolor='RosyBrown1')
e1.grid(row=1,column=1,padx=10,pady=10)
e1=Label(window,text="Metacademy Username")
e1.grid(row=0,column=1,padx=5,pady=5)
e2_value=StringVar()
e2=Entry(window,width=20,textvariable=e2_value)
e2.grid(row=4,column=1,padx=10,pady=10)
e2=Label(window,text="Metacademy Password")
e2.grid(row=3,column=1,padx=5,pady=5)
e3_value=StringVar()
e3=Entry(window,width=56,textvariable=e3_value)
e3.grid(row=7,column=1,padx=10,pady=10)
e3=Label(window,text="Location of file in your system")
e3.grid(row=6,column=1,padx=5,pady=5)

window.mainloop()
```
